```python
import cooler
import matplotlib.pyplot as plt
import numpy as np
```

```python
TASK_SIZE = 100
```

```python
def plot_hic(matrix):
    fig = plt.figure(figsize=(10, 10))
    ax = fig.add_subplot(111)
    im = ax.matshow(np.log10(matrix), cmap='YlOrRd')
    fig.colorbar(im)
```

```python
def mark_nans_as_zeros(matrix):
    matrix[np.isnan(matrix)] = 0
```

```python
# import hi-c data (drosophila embryo)
filepath = "Kc167.10000.cool"
c = cooler.Cooler(filepath)

# TODO (future): train на одной хромосоме, test на другой хромосоме
# получаем карту контактов для одной хромосомы, нам не нужны межхромосомные взаимодействия
arr = c.matrix().fetch("chr2L")

current_matrix = arr[0:TASK_SIZE, 0:TASK_SIZE]
```

```python
plot_hic(current_matrix)
```

```python
bad_row_dataset = current_matrix[52:55, ] # bad row in the middle
good_row_dataset = current_matrix[16:19, ] # good row in the middle

mark_nans_as_zeros(bad_row_dataset)
mark_nans_as_zeros(good_row_dataset)

plot_hic(bad_row_dataset)
plot_hic(good_row_dataset)
```

```python
def scatter_plot(x, y, plot_name):
    plt.scatter(x, y)
    plt.title(plot_name)
    plt.xlabel('Average value of upper and lower pixels')
    plt.ylabel('Value of the middle pixel')
    plt.show()
```

```python
x, y = np.average(np.array([good_row_dataset[0, ], good_row_dataset[2, ]]), axis=0), good_row_dataset[1, ]

# remove 3 outliers
x = np.delete(x, np.argmax(y))
y = np.delete(y, np.argmax(y))

y = np.delete(y, np.argmax(x))
x = np.delete(x, np.argmax(x))

y = np.delete(y, np.argmax(x))
x = np.delete(x, np.argmax(x))

scatter_plot(x, y, 'Good row (>5% non-zero beans) scatter plot')
```

```python
x, y = np.average(np.array([bad_row_dataset[0, ], bad_row_dataset[2, ]]), axis=0), bad_row_dataset[1, ]

# remove 3 outliers
x = np.delete(x, np.argmax(y))
y = np.delete(y, np.argmax(y))

y = np.delete(y, np.argmax(x))
x = np.delete(x, np.argmax(x))

y = np.delete(y, np.argmax(x))
x = np.delete(x, np.argmax(x))

scatter_plot(x, y, 'Bad row (<5% non-zero beans) scatter plot')
```
